# Exploratory Data Analysis (EDA)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# create df
df = pd.read_csv('hr_employee_churn_data.csv')
```

```python
# check total employee range baed on target column 'left'
sns.countplot(x='left', data=df)

# Takeaway: Less people left compared to those who stayed. There exists a data imbalance.
```

```python
# check total employee range based on salary
sns.countplot(x='salary', data=df)

# Takeaway: High count of people with low and medium salaries which is expected in an organization
```

```python
# Split those who left and stayed according to salary
sns.catplot(x='left', col='salary', kind='count', data=df)

#Takeaway: Low salary employees have relatively higher churn rate
```

```python
# check employee distribution based on promotion in the last 5 years
sns.countplot(x='promotion_last_5years', data=df)

#Takeaway: Non-promote employee count is very high
```

```python
# Split those who left and stayed according to the promotions in the last 5 yrs
sns.catplot(x='left', col='promotion_last_5years', kind='count', data=df)

#Takeaway: There is a higher churn rate for people who are not promoted in the last 5 yrs
```

```python
# check employee distribution based on work accident
sns.countplot(x='Work_accident', data=df)

#Takeaway: There are relatively few work accidents in the organization
```

```python
# Split those who left and stayed according to the promotions in the last 5 yrs
sns.catplot(x='left', col='Work_accident', kind='count', data=df)

#Takeaway: Even though the employee does not experience a work accident, the churn rate is high. This feature, at face value, does not seem so important.
```

```python
# create a distribution plot for the satisfaction level of employees
sns.displot(data=df, x='satisfaction_level', kde=True)

#Takeaway: Satisfaction level of employees is distributed from 0 to 1 but has high count at 0.1
```

```python
# create box plot to show distrubutions with respect to categories
sns.boxplot(x='left', y='satisfaction_level', data=df)

#Takeaway: The employees who left have lower satisfaction level compared to those who stayed
```

```python
# create a distribution plot for the number of project
sns.displot(data=df, x='number_project', kde=True)

#Takeaway: The number of projects spans from 2 to 7 and majority of the employees have 3 and 4 projects.
```

```python
# create box plot to show distrubutions with respect to categories
sns.boxplot(x='left', y='number_project', data=df)

#Takeaway: The employees who left have higher number of projects
```

```python
# create a distribution plot for the last evaluation score of the employees
sns.displot(data=df, x='last_evaluation', kde=True)

#Takeaway: The evaluation scores span from 0.36 to 1 with the highest count at around 0.5
```

```python
# create box plot to show distrubutions with respect to last evaluation score
sns.boxplot(x='left', y='last_evaluation', data=df)

#Takeaway: The employees who left have higher evaluation scores 
```

```python
# create a distribution plot for the average monthly hours of the employees
sns.displot(data=df, x='average_montly_hours', kde=True)

#Takeaway: The average monthly hours span from 100 to 300 and it has a high count at around 150
```

```python
# create box plot to show distrubutions with respect to average monthly hours
sns.boxplot(x='left', y='average_montly_hours', data=df)

#Takeaway: The employees who left have higher monthly hours
```

```python

# create a distribution plot for the time spent in the company of the employees
sns.displot(data=df, x='time_spend_company', kde=True)

#Takeaway: The average monthly hours span from 2 to 10 and majority has spent 3 years
```

```python
# create box plot to show distrubutions with respect to time spent in the company
sns.boxplot(x='left', y='time_spend_company', data=df)

#Takeaway: The employees who left have longer time spent in the company
```
